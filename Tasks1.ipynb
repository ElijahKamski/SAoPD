```python
a=list(map(int, input().split()))
a.sort()
ans=[-1,-1,-1]
for i in range(len(a)-2):
    if(a[i+2]<a[i]+a[i+1]):
        ans=a[i:i+3]
print(' '.join(map(str,ans)))
print(sum(ans))
```

```python
from functools import cmp_to_key
def compare(item1, item2):
    s1=item1
    s2=item2
    w1=sum([int(s1[i])/(i+1+len(s1)) for i in range(len(s1))])
    w2=sum([int(s2[i])/(i+1+len(s2)) for i in range(len(s2))])
    return w1-w2
a=input().split()
a.sort(key=cmp_to_key(compare))
print(''.join(a[::-1]))
```
